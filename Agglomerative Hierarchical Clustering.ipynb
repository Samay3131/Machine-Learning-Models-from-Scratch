### Question 5(b) 

Single linkage considers the minimum amongst the euclidean distances of all the datapoints regardless the dissimilarities of the groups which are getting merged togeather.Therfore it will combine the observations linked by a series of close intermidiate observations. This phenomenon is refred to as chaining. Hence the clusters which are very apart from each other can be merged/considered in the same group. This can produce an irregualar dendogram and bias may introduce. 

# Practical Part 

```python
random.seed(3105)
datapoints=[]
point_x=[]
point_y=[]
point_z=[]

for i in range(100):
        point_x.append(random.randrange(-50,50))
        point_y.append( random.randrange(-50,50))
        point_z.append( random.randrange(-50,50))
        point=[point_x[i],point_y[i],point_z[i]]
        datapoints.append(point)
```

```python
def distance(point1, point2):
    dist =  sum([(i - j)**2 for i, j in zip(point1, point2)])
    return pow(dist, 0.5)
```

```python
class Node:
    def __init__(self, x):

        if not isinstance(x[0], list):
            self.cluster = [x]
        else:
            self.cluster = [i for i in x]  
        self.height = 0
        self.leftChild = None
        self.rightChild = None
```

```python
def singleLinkage(node1, node2):
    #single linkage
    dist=[]
    for i in node1.cluster:
        for j in node2.cluster:
            dist.append(distance(i,j))
    return min(dist)

def completeLinkage(node1, node2):
    # complete linkage
    dist=[]
    for i in node1.cluster:
        for j in node2.cluster:
            dist.append(distance(i,j))
    return max(dist)

def averageLinkage(node1, node2):
    # average linkage
    dist=[]
    for i in node1.cluster:
        for j in node2.cluster:
            dist.append(distance(i,j))
    return sum(dist)/len(dist)
    
def centroidLinkage(node1, node2):
    # centroid linkaage
    centroid_1 =  [sum(i)/len(node1.cluster) for i in zip(*node1.cluster)]
    centroid_2 =  [sum(i)/len(node2.cluster) for i in zip(*node2.cluster)]
    dist = distance(centroid_1,centroid_2)
    return dist
    
    
    
```

```python
def clustersToMerge(currentClusters, linkage):
    #clusters to merge
    min=1000
    for i in range(len(currentClusters)):
        for j in range(len(currentClusters)):
            if i != j and j>i:
                if min>linkage( currentClusters[i], currentClusters[j]):
                    min=linkage( currentClusters[i], currentClusters[j])
                    i1,j1,min=i,j,min
    return (i1,j1,min)
```

```python
def ahc(datapoints, linkage):
    # returns the root node of the AHC tree
    currentClusters=[]
    for i in range(len(datapoints)):
        currentClusters.append(Node(datapoints[i]))
        
    while len(currentClusters) >1:
        (i,j,h) = clustersToMerge(currentClusters, linkage)
        Nnode = Node(currentClusters[i].cluster + currentClusters[j].cluster)
        Nnode.height = h
        Nnode.leftChild = currentClusters[i]
        Nnode.rightChild = currentClusters[j]
        NewNodes = []
        
        for idx in range(len(currentClusters)):
            if idx==i or idx==j:
                pass
            else:
                NewNodes.append(currentClusters[idx])
        NewNodes.append(Nnode)
        currentClusters = NewNodes.copy()
    return Nnode
        
            
        
```

```python
def cutTree(rootNode, k):
    # write the content of this function
    # returns the k clusters obtained by cutting the AHC tree at an appropriate height
    current_cluster = rootNode
    cluster = []
    cnt =2
    if current_cluster.leftChild.height > current_cluster.rightChild.height:
        cluster.append(current_cluster.leftChild)
    else :
        cluster.append(current_cluster.rightChild)
    i = 0
    nodes = []  
    while cnt <= k: 
     
     if cluster[i].leftChild or cluster[i].rightChild:
            if cluster[i].leftChild.height > cluster[i].rightChild.height:
                cluster.append(cluster[i].leftChild)
                i=i+1
                cnt = cnt+1
            else:
                cluster.append(cluster[i].rightChild)
                i=i+1
                cnt = cnt+1
    return cluster

# Write your code for generating the plots here
def plot_graph(cut):
    points = []
    fig = plt.figure()
    fig.set_figwidth(15)
    fig.set_figheight(7)
    axis = plt.axes(projection='3d')    
    for i in range(len(cut)):
        points.append(cut[i].cluster)
    for i in range(len(points)): 
        x,y,z = zip(*points[i])
        axis.scatter3D(x, y, z,label = i)
    plt.legend()
    plt.show()
    
        
```

```python

rootNode_single = ahc(datapoints, singleLinkage) #rootnode for single linkage
rootNode_complete = ahc(datapoints, completeLinkage)  #rootnode for complete linkage
rootNode_averageLinkage = ahc(datapoints, averageLinkage) #rootnode for average linkage
rootNode_centroid = ahc(datapoints, centroidLinkage) #rootnode for centroid linkage

```

```python
cutTree_single= cutTree(rootNode_single, 5) #cutree for single linkage
cutTree_complete= cutTree(rootNode_complete, 5)  #cutree for complete linkage
cutTree_averageLinkage= cutTree(rootNode_averageLinkage, 5)  #cutree for average linkage
cutTree_centroid= cutTree(rootNode_centroid, 5) #cutree for centroid linkage

```
